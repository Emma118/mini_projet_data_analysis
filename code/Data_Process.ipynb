```python
import pandas as pd
headers_path = r'C:\Users\GLVC5389\Desktop\Master_DF-IA\Data Process\Mini_projet\header.csv'
data_path = r'C:\Users\GLVC5389\Desktop\Master_DF-IA\Data Process\Mini_projet\BF.csv'


#lecture des entêtes

with open(headers_path, 'r') as file:
    headers = file.readline().split(';')
 
    #Lecture des données

data = pd.read_csv(data_path, header=None, sep=';')

#Recupération des entêtes
data.columns = headers

# traitement des espaces
data.columns = data.columns.str.replace(r"^\s+|\s+$", "", regex=True)

print(data.columns)

```

```python
new_comlumns = ['geonameid', 'name', 'latitude', 'longitude']

new_data = data[new_comlumns]

new_columns = {'geonameid': 'ID', 'name': 'location_name', 'latitude': 'latitudes', 'longitude': 'longitudes'}

new_data_renamed = new_data.rename(columns=new_columns)

new_data_renamed
```

```python
#Enregistrement du csv
new_data_renamed.to_csv(r'C:\Users\GLVC5389\Desktop\Master_DF-IA\Data Process\Mini_projet\burkina_location.csv', index=False)
```

```python
# Trouver les latitudes et longitudes minimales et afficher les noms correspondant

#Recupérer les index min

index_latitude_min = new_data_renamed['latitudes'].idxmin()
index_longitude_min = new_data_renamed['longitudes'].idxmin()

# Recuperer les valeurs min

latitude_min = new_data_renamed.loc[index_latitude_min, 'latitudes']
longitude_min = new_data_renamed.loc[index_longitude_min, 'longitudes']

# Recuperer les noms

nom_latitude_min = new_data_renamed.loc[index_latitude_min, 'location_name']
nom_longitude_min = new_data_renamed.loc[index_longitude_min, 'location_name']

# Afficher les résultats
print(f"Latitude minimale: {latitude_min}, Lieu: {nom_latitude_min}")
print(f"Longitude minimale: {longitude_min}, Lieu: {nom_longitude_min}")


```
